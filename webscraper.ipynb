```python
# import the requests package
import requests

# import BeautifulSoup

from bs4 import BeautifulSoup
```

```python
url = "https://www.ncwildlife.org/Learning/Species#8682115-mollusks"
data = requests.get(url)
data
```

```python
html = data.text
html
```

Parsing Data

```python
soup = BeautifulSoup(html, "html.parser")
print(soup.prettify)
```

```python
mollusks = soup.find(id="la-8682115-mollusks")
mollusks
```

```python
column_headers = ["Name", "Link"]

final_data_list = []

links = mollusks.find_all("a")

for i in range(5, len(links)):
    row_data = {}
    row_data["Name"] = links[i].text
    row_data["Link"] = links[i].get("href")
    final_data_list.append(row_data)

final_data_list
```

```python
import pandas as pd

df = pd.DataFrame.from_dict(final_data_list)

df.head()
```
